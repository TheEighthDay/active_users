```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
```

```python
app_launch = pd.read_csv("../original_data/app_launch_log.txt", sep='\t', names=['user_id', 'day'])
```

```python
user_register = pd.read_csv("../original_data/user_register_log.txt", sep='\t', names=['user_id', 'register_day', 'register_type', 'device_type'])
```

```python
video_create = pd.read_csv("../original_data/video_create_log.txt", sep='\t', names=['user_id', 'day'])
```

```python
user_activity = pd.read_csv("../original_data/user_activity_log.txt", sep="\t", names=['user_id', 'day', 'page', 'video_id', 'author_id', 'action_type'])
```

## 生成label

```python
test_data = app_launch[app_launch['day']>=24]
```

```python
positive_user = test_data[['user_id']].drop_duplicates()
```

```python
positive_user['flag'] = 1
```

```python
label = pd.merge(positive_user, user_register[['user_id']], on='user_id', how='outer').fillna(0)
```

```python
label[label['flag'] == 1].count()
```

```python
label[label['flag'] == 0].count()
```

```python
user_register_7_ago = user_register[user_register['register_day']<24][['user_id']]
```

```python
label = pd.merge(user_register_7_ago, label, on='user_id', how='left')
```

```python
label[label['flag'] == 1].count()
```

```python
label[label['flag'] == 0].count()
```

```python
label.iloc[[1,23]]
```
